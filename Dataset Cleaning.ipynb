```python
# drop duplicate and unused columns
sales.drop("Unnamed: 0", inplace=True, axis=1)
sales.drop("EASE-MENT", inplace=True, axis=1)
sales.drop("APARTMENT NUMBER", inplace=True, axis=1)
sales.drop("newDate_x", inplace=True, axis=1)
sales.drop("newDate_y", inplace=True, axis=1)
```

```python
sales.rename(columns= {"newValue_x":'30 Year Rate',"newValue_y":'15 Year Rate'},inplace=True)
```

```python
# calculate age of property at time of sale
sales["SALE DATE"] = pd.to_datetime(sales["SALE DATE"])
age = (sales["SALE DATE"].dt.year - sales["YEAR BUILT"])
for i in range(len(age)):
    if age[i] > 2000:
        age[i] = np.nan

sales["AGE"] = age
sales.head(100)
```
